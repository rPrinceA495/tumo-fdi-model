```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

'''
Outer join of dataframes on TumoID
'''
# List of file locations
# files = ['socioeconomic and demographic/Socioeconomic Aggregation.xlsx', 'socioeconomic and demographic/Closed Students.xlsx', 'socioeconomic and demographic/Path Date - end of Initiation.xlsx', 'pathology/Student Pathology Scores.xlsx']


files = ['../data/socioeconomic and demographic/Socioeconomic Aggregation.xlsx', '../data/socioeconomic and demographic/Path Date - end of Initiation.xlsx', '../data/pathology/Student Pathology Scores.xlsx', '../data/pedagogic engagement/self-learning/ready/self_learning_data.xlsx', '../data/pedagogic engagement/workshops and masterlabs/ready/workshop_masterlab_data.xlsx']
# Read the first file
df = pd.read_excel(files[0])

# Loop over the rest of the files and merge
for file in files[1:]:
    df_other = pd.read_excel(file)
    df = pd.merge(df, df_other, on='TumoID', how='outer')
    
df.head()
# Check for missing values in the dataframe 
# df.isnull().sum()
```

```python
# Add missing pathology and Socioeconomic Scores
df['PHDisability'] = df['PHDisability'].fillna(0)
df['PHChronicDisease'] = df['PHChronicDisease'].fillna(0)
df['PHDevelopmentalCoordination'] = df['PHDevelopmentalCoordination'].fillna(0)
df['MHNeurodevelopmental'] = df['MHNeurodevelopmental'].fillna(0)
df['MHSpeechCommuication'] = df['MHSpeechCommuication'].fillna(0)
df['MHEmotional'] = df['MHEmotional'].fillna(0)
df['MHPsychiatric'] = df['MHPsychiatric'].fillna(0)
df['MHFamilyEnvironment'] = df['MHFamilyEnvironment'].fillna(0)
df['MHSocialBehavioral'] = df['MHSocialBehavioral'].fillna(0)
df['MHLearningSchool'] = df['MHLearningSchool'].fillna(0)
df['PathologyScore'] = df['PathologyScore'].fillna(0)

df.head()

# Check for missing values in the dataframe 
# df.isnull().sum()

# Check for missing values in the dataframe
# df.isnull().sum() 
```

```python
# Add target value (label)
df['DroppedOut'] = df.apply(lambda row: 1 if row['Status'] == 'Closed' else 0, axis=1)

df.head()

# Drop the rows with null values
df = df[df['SocioeconomicScore'].notna()]
```

```python
# Check for missing values in the dataframe
df.isnull().sum() 

# Specify the columns you want to drop
columns_to_drop = ['TumoID','AccountClosingDate', 'TerminationType', 'TerminationReason', 'Remarque', 'Noticed', 'ApplicationDate', 'RegistrationDate','PostalCode', 'TumoStatus', 'StudentSchedule', 'Status', 'Age', 'RetentionByMonths', 'PathDate', 'AttendingSince']

# Drop the columns
df = df.drop(columns=columns_to_drop)

```

```python
# Check for missing values in the dataframe (None found)
df.isnull().sum() 

# Drop the rows with null values
df = df.dropna()
```

```python
# Check for missing values in the dataframe
df.isnull().sum() 

# Replace categorical values in 'Sex' column to 0s and 1s
df['Sex'] = df['Sex'].replace({'Femme': 0, 'Homme': 1, 'Autre': 2})

# Save the merged DataFrame to a new Excel file
df.to_excel('data.xlsx', index=False)
```
