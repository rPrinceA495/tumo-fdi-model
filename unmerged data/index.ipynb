```python
import pandas as pd

# Load self-learning datasets
df_self_learning = pd.read_excel('Self Learning Activities.xlsx')
df_self_learning.head()
```

```python
# Column Recoding (TryStatus)

# Define a dictionary that maps the current try status values to the new numerical values
status_dict = {'New': 0, 'Working': 1, 'Pending': 2, 'Rejected': 1, 'Awarded': 3}

# Use the replace function to recode the status values and store them in a new column
df_self_learning['TryStatusCode'] = df_self_learning['TryStatus'].replace(status_dict)

df_self_learning.head()
```

```python
# Define a function to calculate mode
def mode(x):
    return x.mode()[0] if len(x.mode()) > 0 else None

# Define a function to count attempted activities
def count_attempted(x):
    return (x != 'New').sum()

# Aggregate data
df_self_learning_agg = df_self_learning.groupby('TumoID').agg({
    'TryStatus': count_attempted,  # Count
    'RejectCount': 'sum',  # Total number of attempts
    'TryStatusCode': ['mean', 'median', mode]  # Mean, median, and mode of status code
})

# Flatten the multi-level column index
df_self_learning_agg.columns = ['_'.join(col).strip() for col in df_self_learning_agg.columns.values]

# Rename the count column
df_self_learning_agg.rename(columns={'TryStatus_count_attempted': 'TotalActivitiesAttempted'}, inplace=True)

df_self_learning_agg.head()

# Export to Excel
df_self_learning_agg.to_excel('aggregated_data.xlsx')
```
