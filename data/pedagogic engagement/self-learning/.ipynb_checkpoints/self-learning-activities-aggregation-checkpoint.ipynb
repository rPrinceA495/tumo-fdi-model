```python
import pandas as pd

# Load self-learning datasets
df_self_learning = pd.read_excel('Self Learning Activities.xlsx')
df_self_learning.head()
```

```python
# Group by 'Student ID' and 'Try Status', then count the number of activities in each status
status_counts = df_self_learning.groupby('TumoID')['TryStatus'].value_counts().unstack(fill_value=0)

# Define a function to count non-'New' activities
def count_non_new(x):
    return (x != 'New').sum()

# Add a new column 'Non-New Activities' to the DataFrame
status_counts['NumOfAttemptedActivities'] = df_self_learning.groupby('TumoID')['TryStatus'].apply(count_non_new)

# Now, 'Non-New Activities' is a new column in your DataFrame that contains the count of activities that are not 'New' for each student.

# To export this DataFrame to an Excel file, you can use the to_excel function:
status_counts.to_excel('Self Learning Activities Aggregation.xlsx')


# status_counts.head()
```

```python
import matplotlib.pyplot as plt
# Visualize Activity Status
# Summarize the data
columns = ['New', 'Pending', 'Rejected', 'Awarded','Working']
# Summarize the data
summary = status_counts[columns].sum()

# Calculate percentages
percentages = 100. * summary / summary.sum()

# Create labels with percentages
labels = ['{0} - {1:1.1f} %'.format(i,j) for i,j in zip(summary.index, percentages)]

# Plotting
plt.figure(figsize=(10,7))
plt.pie(summary, labels=labels, autopct='%1.1f%%')
plt.title('Distribution of Selected Activity Statuses')
plt.show()
```

```python
# Bar Plot
summary.plot(kind='bar', figsize=(10,7))
plt.title('Distribution of Activity Statuses')
plt.xlabel('Activity Status')
plt.ylabel('Number of Activities')
plt.show()
```

```python
'''
Outer join of dataframes on TumoID
'''
# List of file locations
files = ['Attendance - Self Learning.xlsx', 'Closed Students.xlsx', 'Continuous Absence in Self Learning session since the last log in.xlsx', 'Path Date - end of Initiation.xlsx', 'Self Learning Activities Aggregation.xlsx']

# Read the first file
df = pd.read_excel(files[0])

# Loop over the rest of the files and merge
for file in files[1:]:
    df_other = pd.read_excel(file)
    df = pd.merge(df, df_other, on='TumoID', how='outer')

# Save the merged DataFrame to a new Excel file
df.to_excel('ready/self_learning_data.xlsx', index=False)
```
